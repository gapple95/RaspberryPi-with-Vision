# 01. 변수란 무엇인가?


image.png


변수는 데이터를 임시로 저장하는 **저장소**입니다.  
주로 숫자 또는 문자를 저장하며, 프로그램 실행 중에 값이 **변경**될 수 있습니다.

예를 들어, 게임에서 다음과 같은 데이터를 변수로 관리할 수 있습니다:
- **점수(Score)**: 플레이어가 획득한 점수
- **레벨(Level)**: 현재 플레이어의 진행 상태
- **남은 시간(Time)**: 게임이 종료되기까지의 시간


## 01-2 변수의 주요 특징

1. **임시 저장소**: 데이터를 저장하고 필요할 때 값을 참조하거나 수정할 수 있습니다.
2. **변경 가능**: 변수에 저장된 값은 프로그램의 실행 중에도 자유롭게 변경될 수 있습니다.
3. **유연성**: 숫자, 문자 등 다양한 데이터를 저장할 수 있습니다.

---

### 1. 점수 관리
- 플레이어가 획득한 점수를 실시간으로 저장하고 업데이트.
- **예제**:  
  ```python
  score = 0  # 초기 점수

# 02. 변수 선언(Declaring a Variable)


image.png


- **변수 선언**은 프로그램에서 특정 데이터를 저장하기 위해 이름과 데이터 타입을 지정하는 과정입니다.
- 예를 들어, 변수 이름을 `score`로 지정하고 데이터 타입을 정수(`int`)로 설정합니다.

## 02-1. 변수 선언의 구성 요소

1. **변수 이름 (Variable Name)**:
   - 데이터를 저장하는 공간의 **이름**입니다.
   - 예: `score`
   - 변수 이름의 규칙:
     - 알파벳, 숫자, 밑줄(_)을 사용할 수 있습니다.
     - 숫자로 시작할 수 없습니다.
     - 예약어(예: `int`, `for`, `while` 등)는 사용할 수 없습니다.

2. **데이터 타입 (Data Type)**:
   - 변수에 저장할 데이터의 **종류**를 지정합니다.
   - 주요 데이터 타입:
     - **정수 (int)**: 숫자 데이터를 저장.
     - **실수 (float)**: 소수점이 있는 숫자 데이터를 저장.
     - **문자열 (str)**: 텍스트 데이터를 저장.
     - **불리언 (bool)**: `True` 또는 `False` 값을 저장.

## 02-3. 변수 선언 예제

```python
# 변수 선언
score = 0  # 변수 이름: score, 데이터 타입: int
level = 1  # 변수 이름: level, 데이터 타입: int
player_name = "Player1"  # 변수 이름: player_name, 데이터 타입: str
is_game_over = False  # 변수 이름: is_game_over, 데이터 타입: bool

# 출력 예제
print("점수:", score)
print("레벨:", level)
print("플레이어 이름:", player_name)
print("게임 종료 여부:", is_game_over)



# 03. 변수 초기화(Initialising a Variable)


image.png

- 변수 초기화는 변수에 **처음 값**을 할당하는 것을 의미합니다.
- 프로그램 실행 중 변수는 반드시 초기화되어야 올바르게 작동합니다.
- 예: 변수 `score`를 초기화하여 0의 값을 할당합니다.








## 03-2. 변수 초기화 방법

1. **값 할당 (Assignment)**:
   - 변수에 특정 값을 설정하여 초기화합니다.
   - 예: `score = 0`

2. **초기값 설정 이유**:
   - 초기값을 설정하지 않으면 변수는 비어 있는 상태로, 프로그램 실행 시 오류가 발생할 수 있습니다.

## 03-3. 변수 초기화 예제

```python
# 변수 초기화
score = 0  # 점수 변수 초기화
level = 1  # 레벨 변수 초기화

# 초기화된 변수 출력
print("초기 점수:", score)
print("초기 레벨:", level)
```

# 04. 변수에 값 할당(Assigning a Value to a Variable)


image.png

- 변수에 새로운 값을 저장하거나 기존 값을 업데이트하는 작업입니다.
- 예: `score` 변수에 10점을 더하거나, 기존 값에서 업데이트하여 20점을 저장합니다.





## 04-2. 변수 값 할당 방법

1. **새로운 값 할당**:
   - 변수를 선언하고 값을 할당합니다.
   - 예: `score = 10`

2. **값 업데이트**:
   - 기존 변수에 새로운 값을 추가하거나 수정합니다.
   - 예: `score = score + 10` 또는 `score += 10`




## 04-3. 변수 값 할당 예제

```python
# 변수 선언 및 초기화
score = 10  # 점수 변수에 10 할당
print("현재 점수:", score)

# 값 업데이트
score += 10  # 기존 점수에 10 추가
print("업데이트된 점수:", score)

# 새로운 값으로 재할당
score = 20  # 점수를 20으로 업데이트
print("최종 점수:", score)
```

# 05. 데이터 타입(Data Types)


image.png

- 데이터 타입은 변수에 저장할 **값의 종류**를 정의합니다.
- 프로그램이 데이터를 효과적으로 처리하기 위해 각 데이터에 맞는 타입을 사용해야 합니다.
- 데이터 타입은 변수의 **크기**와 **사용 목적**에 따라 선택됩니다.


## 05-2. 주요 데이터 타입

1. **정수형 (int)**:
   - 숫자 데이터를 저장합니다.
   - 예: `level = 1` (레벨 데이터)
   
2. **문자열형 (str)**:
   - 문자나 텍스트 데이터를 저장합니다.
   - 예: `name = "Jonathan"` (플레이어 이름)

3. **불리언형 (bool)**:
   - 참(True) 또는 거짓(False)의 값을 저장합니다.
   - 예: `sound = True` (사운드 상태)




````python
#1. 데이터 타입 예제

# 정수형 예제
level = 1  # 플레이어 레벨
print("레벨:", level)

# 실수형 예제
angle = 97.722
print("천사:", angle)

# 문자열형 예제
name = "Jonathan"  # 플레이어 이름
print("플레이어 이름:", name)

# 불리언형 예제
sound = True  # 사운드 상태
print("사운드 활성화 여부:", sound)
````
